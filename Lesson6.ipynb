#### Задание 1

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.keys()
```

```python
feature_names = boston['feature_names']
feature_names
```

```python
data = boston ['data']
target = boston ['target']
```

```python
X= pd. DataFrame(data, columns = feature_names)
X.head(10)
```

```python
y = pd.DataFrame(target, columns = ['price'])
y.head(10)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred_lr = lr.predict(X_test)
```

```python
check_test = pd.DataFrame({'y_test': y_test['price'],
                         'y_pred_lr': y_pred_lr.flatten()})
check_test.head()
```

```python
#находим коэфф детерминации
from sklearn.metrics import r2_score
```

```python
lr_r2_score=r2_score(check_test['y_test'], check_test['y_pred_lr'])
lr_r2_score
```

```python
df = pd.read_csv('creditcard.csv')
df.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
df['Class'].value_counts(normalize = True)
```

```python
df.info()
```

```python
pd.options.display.max_columns = 100
```

```python
df.head(10)
```

```python
X = df.drop('Class', axis = 1)
y = df['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100, stratify = y)
```

```python
print('X_train ', X_train.shape)
print('X_test ', X_test.shape)
print('y_train ', y_train.shape)
print('y_test ', y_test.shape)
```

```python
parameters = [{'n_estimators': [10, 15],
'max_features': np.arange(3, 5),
'max_depth': np.arange(4, 7)}]
```

```python
clf = GridSearchCV(
    estimator=RandomForestClassifier(random_state = 100),
    param_grid = parameters,
    scoring = 'roc_auc',
    cv = 3,)
```

```python
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
y_predict_proba = clf.predict_proba(X_test)
```

```python
y_predict_proba[:10]
```

```python
y_pred_proba = y_predict_proba[:, 1]
```

```python
y_pred_proba[:10]
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_test, y_pred_proba)
```

```python
# 8. Постройте матрицу корреляций для всех полей X. Дайте полученному датафрейму название X_corr.
X_corr = X.corr()
X_corr = np.round(X_corr, 2)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10, 5))
sns.heatmap(X_corr, annot = True, cmap = 'coolwarm')
```

```python
# 9. Создайте список high_corr из признаков, корреляция которых с полем target по абсолютному значению превышает 0.5
# (причем, само поле target не должно входить в этот список).

high_corr = X_corr.loc[(np.abs(X_corr['target']) > 0.5) & (X_corr.index != 'target'), X_corr.columns != 'target'].index
high_corr
```
